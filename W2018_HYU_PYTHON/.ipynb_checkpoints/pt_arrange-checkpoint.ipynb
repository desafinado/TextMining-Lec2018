```python
import nltk
#nltk.download('maxent_treebank_pos_tagger') # 다운로드된 폴더를 옮겨온다.
#nltk.download('averaged_perceptron_tagger')
```

### 3. Phrase Chunking - 품사들을 결합하여 별도로 인식하기

***

### 3-1. 명사만 if (t[1]=='NN') 뽑아서 데이터프레임으로 만든다.

```python
tok_noun = [t for d in doc_all for t in d  if (t[1]=='NN' and len(t[0]) > 1) ]
print('전체명사개수:', len(tok_noun), ' ' ,'중복을 제거한 명사 개수 :',len(set(tok_noun)))
```

```python
text_all = nltk.Text(tok_all,name="NMSC")
text_noun = nltk.Text(tok_noun,name="NMSC")
```

```python
# 명사를 Count해서 excel로 기록한다.
import pandas as pd
freq_nu = text_noun.vocab().most_common(len(set(text_noun.tokens)))
freq_nu_cl = [(freq_nu[i][0][0] , freq_nu[i][1]) for i in range(len(freq_nu))]
freq_NN1 = pd.DataFrame(freq_nu_cl)
freq_NN1.rename(columns={0:'Var', 1 : 'Count'}, inplace=True)
freq_NN1['P'] = "Noun"
#freq_NN1.to_excel("output/test.xlsx")
```

### 3-2. 명사+명사 연어(collocation)을 뽑아낸다.
##### 위의 "명사+명사" 이외로 뽑고 싶다면 grammar = "2NN : {NN NN}" 코드를 수정한다.

```python
grammar = "2NN : {<NN><NN>}"
cp = nltk.RegexpParser(grammar)
a= cp.parse([t for d in doc_all for t in d])
NN2 = list(a.subtrees(filter=lambda x: x.label() == "2NN" ))
NN2_list = [NN2[i][0] + NN2[i][1] for i in range(len(NN2))]
NN2_frame = pd.DataFrame(NN2_list)
NN2_frame['NN2'] = NN2_frame[0]+' '+NN2_frame[2]
freq_NN2 = (NN2_frame['NN2'].value_counts()).to_frame()
freq_NN2['Var'] = freq_NN2.index
freq_NN2['Count'] = freq_NN2['NN2']
freq_NN2_cl = freq_NN2.reset_index()
freq_NN2_cl.drop(['index','NN2'], inplace=True, axis=1)
freq_NN2_cl['P'] = "NN"
```

### 3-3. 형용사+명사 연어(collocation)을 뽑아낸다.

```python
grammar = "2JN : {<JJ><NN>}"
cp = nltk.RegexpParser(grammar)
a= cp.parse([t for d in doc_all for t in d])
JN = list(a.subtrees(filter=lambda x: x.label() == "2JN" ))
JN_list = [JN[i][0] + JN[i][1] for i in range(len(JN))]
JN_frame = pd.DataFrame(JN_list)
JN_frame['JN'] = JN_frame[0]+' '+JN_frame[2]
freq_JN = (JN_frame['JN'].value_counts()).to_frame()
freq_JN['Var'] = freq_JN.index
freq_JN['Count'] = freq_JN['JN']
freq_JN_cl = freq_JN.reset_index()
freq_JN_cl.drop(['index','JN'], inplace=True, axis=1)
freq_JN_cl['P'] = "JN"
```

#### 명사의 빈도가 높음으로 명사+명사, 형용사+명사의 비중을 조금 더 높여서 250개의 Keyword를 선택한다

```python
freq_all = N1[:50].append(JN[:100].append(N2[:100])).sort_values(by = "Count", ascending=False)
```

```python
top_250 = dt['Var'].tolist()
```

#### space(공백)이 있으면 처리에 불편함이 발생하여 '-' 처리한다.

```python
kw_ch = [kw.replace(' ','-')  for kw in top_250]
ch_map = [ {kw_ch[i] : top_250[i]} for i in range(len(kw_ch)) ]
```

```python
class TextReplacer():
    def __init__(self, text, lod):
        self.text = text
        self.iterate(lod)
    def replace_kv(self, _dict):
        """Replace any occurrence of a value with the key"""
        for key, value in _dict.items():
            label = """{0}""".format(key)
            self.text = self.text.replace(value, label)
            return self.text
    def iterate(self, lod):
        """Iterate over each dict object in a given list of dicts, `lod` """
        for _dict in lod:
            self.text = self.replace_kv(_dict)
        return self.text
```

#### 원문에서 Keyword에 space(공백)이 있으면 '-' 처리한다.

```python
lst_ch = [TextReplacer(lst_i ,ch_map ).text for lst_i in lst]
```

### 5. Graph 자료구조로 변환 - networkx 라이브러리

***

#### 1개의 문서에 동시에 등장하는 단어를 [ [ "A", "B", "C" ] , ["A", "B"] , ["C", "D", "E", "F"] ] list 형태로 수정하고 같이 등장하면 edge로 연결한다.

```python
import networkx as nx
from itertools import product
to_graph_tmp = [' '.join([x for x in lst_i.split() if x in kw_ch  ]) for lst_i in lst_ch ]
to_graph = [ to_graph_tmp_i for to_graph_tmp_i in to_graph_tmp if to_graph_tmp_i not in [''] ]

to_graph_edges = list(product(to_graph, to_graph))
G = nx.Graph(nodetype = str) 
G.add_edges_from(to_graph_edges)
G.remove_edges_from(G.selfloop_edges())
```

```python
raw_source_kw_tmp = pd.merge(left=kw_freq_all,right=kw_freq_pt, how = 'left', left_index=True, right_index=True)
raw_source_kw = pd.merge(left=raw_source_kw_tmp,right=kw_freq_pa, how = 'left', left_index=True, right_index=True)
```

#### v) 빈도 조건에 맞춰서 원천을 구분한다.

```python
import numpy as np
def raw_source(c):
    if c['freq_pt'] > 0 and np.isnan(c['freq_pa'])  :
        return 'pt'
    elif np.isnan(c['freq_pt']) and c['freq_pa'] > 0 :
        return 'pa'
    elif c['freq_pt'] > 0 and c['freq_pa'] > 0 :
        return 'bo'
```

```python
raw_source_kw['raw_source'] = raw_source_kw.apply(raw_source,axis=1)
raw_source_kw['raw_source'].value_counts()
```

```python
raw_source_dic = dict(zip(raw_source_kw.index ,raw_source_kw['raw_source']))
len(raw_source_dic)
```

#### vi) 조건에 맞춰서 node color를 부여한다.

```python
for k,v in raw_source_dic.items():
    if v == 'pt' :
        G.node[k]['viz'] = {'color':{'r':255, 'g':192 , 'b':0 , 'a':0}} # set node color : RGB value
    if v == 'pa' :
        G.node[k]['viz'] = {'color':{'r':146, 'g':208 , 'b':80 , 'a':0}} # set node color : RGB value
    if v == 'bo' :
        G.node[k]['viz'] = {'color':{'r':141, 'g':180 , 'b':255 , 'a':0}} # set node color : RGB value
```
